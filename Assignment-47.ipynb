```python
#2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

#Ans

#To determine the distribution of flight prices in the dataset, you can compute some descriptive statistics such as the mean, median, and standard deviation. These statistics can provide insights into the central tendency and spread of the flight prices. Here's an example of how you can calculate these statistics using Pandas: 

import pandas as pd

# Load the dataset into a pandas DataFrame
df = pd.read_excel('flight_price.xlsx')

print(df.columns)

flight_prices = df['Price']

# Calculate descriptive statistics
mean_price = flight_prices.mean()
median_price = flight_prices.median()
std_deviation = flight_prices.std()

# Print the descriptive statistics
print("Mean Price:", mean_price)
print("Median Price:", median_price)
print("Standard Deviation:", std_deviation)
```

```python
#4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

#Ans

import pandas as pd
import matplotlib.pyplot as plt

# Load the flight price dataset
flight_data = pd.read_excel('flight_price.xlsx')

# Extract the 'Price' and 'Airline' columns
flight_prices = flight_data['Price']
airlines = flight_data['Airline']

# Create a DataFrame with 'Airline' and 'Price' columns
data = pd.DataFrame({'Airline': airlines, 'Price': flight_prices})

# Create a boxplot to compare the prices of different airlines
plt.figure(figsize=(12, 6))
plt.boxplot(data.groupby('Airline')['Price'].apply(list), labels=data['Airline'].unique())
plt.xlabel('Airline')
plt.ylabel('Flight Price')
plt.title('Flight Prices by Airline')
plt.xticks(rotation=90)
plt.show()
```

```python
#13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

#Ans

import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame with the provided data
data = {
    'App': ['App 1', 'App 2', 'App 3', 'App 4', 'App 5'],
    'Category': ['Games', 'Education', 'Health & Fitness', 'Games', 'Education'],
    'Type': ['Paid', 'Free', 'Paid', 'Free', 'Paid'],
    'Price': ['$2.99', '0', '$4.99', '0', '$1.99']
}
df = pd.DataFrame(data)

# Clean the 'Price' column by removing the '$' sign and converting it to a numeric type
df['Price'] = df['Price'].str.replace('$', '').astype(float)

# Group the dataset by 'Type' and calculate the average price
average_prices = df.groupby('Type')['Price'].mean()

# Create a bar chart
plt.bar(average_prices.index, average_prices)
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')
plt.show()
```

```python
#14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

#Ans

import pandas as pd

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Create a frequency table of app installs
install_freq = df.groupby('App')['Installs'].sum().reset_index()

# Sort the frequency table by install count in descending order
install_freq = install_freq.sort_values(by='Installs', ascending=False)

# Select the top 10 rows
top_10_apps = install_freq.head(10)

# Display the top 10 most popular apps
print(top_10_apps)
```
